از کتابخونه ریجکس استفاده شود برای پیش پردازش

```python
x_subsample = [re.sub(r"[\w\.-]+@[\w\.-]+", "", d) for d in x_subsample]  # Remove email
################# change it
# x_subsample = [re.sub(r"", "", d) for d in x_subsample]  # Remove names
x_subsample = [d.replace("From: ", "") for d in x_subsample]  # Remove "From: "
x_subsample = [
    d.replace("\nSubject: ", "") for d in x_subsample
]
```

```python

df = pd.DataFrame()
df["text"] = x_subsample
df["target"] = list(y_subsample)
```

دکوریت کرده: یعنی تابعی که تکست رو میگیره و امبد شدهشو میده

```python
def make_embed_text_fn(embedding):
    @retry.Retry(timeout=300.0)
    def embed_fn(text):
        return embedding.embed_query(text)

    return embed_fn
```

حالا میخوایم این تابع رو برای ستون تکست اعمال کنیم و برای هر سزش

به مقادیر یک ستون یک فانکشن رو اپلای می کنیم

```python
df["embeddings"] = df["text"].apply(make_embed_text_fn(embedding))
df.head()
```

از روش tsne  برای نشان دادن فضای مساله استفاده میشه یا نمایی ا مساله رو
بدانیم

این کد برای کاهش ابعاد بردارهای embedding با استفاده از الگوریتم t-SNE (t-Distributed Stochastic Neighbor Embedding) استفاده می‌شود. هدف اصلی این کد این است که بردارهای embedding را که معمولاً در فضای برداری با ابعاد بالا (مثل 768 بعد در BERT یا GPT) قرار دارند، به یک فضای دو یا سه بعدی کاهش دهد تا بتوان آن‌ها را به شکل بصری (مثلاً با یک نمودار) نمایش داد. این فرآیند برای تحلیل داده‌های برداری در فضای کوچک‌تر و درک ساختار داده‌ها در سطح بصری مفید است.

الان جدولی خواهیم داشت که یه ستون آن امبدینگ هست و امبدینگ شده متنه و حدود همون ابعاد مدل ابعاد داره و یه ستون تارگت، حالا برای کاهش ابعاد اینو که بالا گفتیم انجام میدیم و احتمالا دو بعدی داری از اون همه ابعاد امبدینگ و ستون تارگت

```python
embeddings_array = np.array(df["embeddings"].to_list(), dtype=np.float32)
tsne = TSNE(random_state=0, max_iter=1000)
tsne_results = tsne.fit_transform(embeddings_array)
```

```python
df_tsne = pd.DataFrame(tsne_results, columns=["TSNE1", "TSNE2"])
df_tsne["target"] = df["target"]  # Add labels column from df_train to df_tsne
df_tsne.head()
```

حالا فضای دو بعدی داریم که ابعاد امبدینگ هستند و هر جمله رو برده به یه مکان، حالا اونایی که هم مکان هستند احتمالا هم دسته اند و اینجا بررسی میشه که آیا اینکه قبلا گفته بودیم هر خبری چه دسته ای داره با چیزی که ترین شده یکی هست با نه

```python
fig, ax = plt.subplots(figsize=(8, 6))  # Set figsize
sns.set_style("darkgrid", {"grid.color": ".6", "grid.linestyle": ":"})
sns.scatterplot(data=df_tsne, x="TSNE1", y="TSNE2", hue="target", palette="hls")
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
plt.title("Scatter plot of news using t-SNE")
plt.xlabel("TSNE1")
plt.ylabel("TSNE2")
plt.axis("equal")
```

مثالی برای کلسیفیکیشن

```python
import pandas as pd
import numpy as np
from ast import literal_eval ### برای تبدیل استرینگ به لیست

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score

datafile_path = "/content/fine_food_reviews_with_embeddings_1k.csv"
df_food = pd.read_csv(datafile_path)
# df_food.head()
```

```python
df_food.head()
```

```python
df_food["embedding"] = df_food.embedding.apply(literal_eval).apply(np.array)  # convert string to array
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    list(df_food.embedding.values), df_food.Score, test_size=0.2, random_state=42
)

# train random forest classifier
clf = RandomForestClassifier(n_estimators=100)
clf.fit(X_train, y_train)
preds = clf.predict(X_test)
probas = clf.predict_proba(X_test)

report = classification_report(y_test, preds)
print(report)
```
